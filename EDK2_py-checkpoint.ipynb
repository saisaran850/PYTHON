```python
import matplotlib.pyplot as pwd
labels = ["Analog Signals","Robotics","Generative AI"]
sizes = [60,90,120]                                                            #matplotlib using pie chart
pwd.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=80)
pwd.title("Pie Chart")
pwd.show()
```

```python
import matplotlib.pyplot as pwd
A = [1,2,3,4,5]
D1,D2 = [11,21,31,41,51],[1101,1201,1301,1401,1501]
pwd.fill_between(A,D1,D2,color = "Blue",alpha = 0.9,label = 'Area 1-2')  #Importing Matplotlib to show Area Plot
pwd.plot(A,D1,label='Line 1',marker='o')
pwd.plot(A,D2,label= 'Line 2',marker='o')
pwd.xlabel("A-axis"),pwd.ylabel("D-axis"),pwd.title("Area Plot")
pwd.legend(),pwd.show()
```
